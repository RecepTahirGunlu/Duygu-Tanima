```python
import os
import shutil

test_dir = 'validation'
target_dir = 'test'

# Hedef dizindeki klasörleri kontrol et
for folder_name in os.listdir(target_dir):
    folder_path = os.path.join(target_dir, folder_name)
    if os.path.isdir(folder_path):
        # Eğer klasörse
        for filename in os.listdir(folder_path):
            if filename.startswith('Pu'):
                # 'PublicTest' olan resimleri bul
                src_path = os.path.join(folder_path, filename)
                dest_folder = os.path.join(test_dir, folder_name)
                os.makedirs(dest_folder, exist_ok=True)
                dest_path = os.path.join(dest_folder, filename)
                shutil.move(src_path, dest_path)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

test_dir = 'test'
validation_dir = 'validation'

class_names = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

fig, axes = plt.subplots(nrows=2, ncols=7, figsize=(15, 5))

# Test klasöründeki örnek resimleri çiz
for i, folder_name in enumerate(os.listdir(test_dir)):
    folder_path = os.path.join(test_dir, folder_name)
    if os.path.isdir(folder_path):
        for j, filename in enumerate(os.listdir(folder_path)[:1]):  # Sadece bir örnek resim al
            img_path = os.path.join(folder_path, filename)
            img = mpimg.imread(img_path)
            axes[0, i].imshow(img, cmap='gray')
            axes[0, i].axis('off')
            axes[0, i].set_title('Test - ' + class_names[i])

# Validation klasöründeki örnek resimleri çiz
for i, folder_name in enumerate(os.listdir(validation_dir)):
    folder_path = os.path.join(validation_dir, folder_name)
    if os.path.isdir(folder_path):
        for j, filename in enumerate(os.listdir(folder_path)[:1]):  # Sadece bir örnek resim al
            img_path = os.path.join(folder_path, filename)
            img = mpimg.imread(img_path)
            axes[1, i].imshow(img, cmap='gray')
            axes[1, i].axis('off')
            axes[1, i].set_title('Validation - ' + class_names[i])

plt.tight_layout()
plt.show()
```
